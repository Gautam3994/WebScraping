```python
import re
import requests

response = requests.get("https://www.summet.com/dmsi/html/codesamples/addresses.html")
html = response.text
print(html)
```

```python
pattern = re.compile(r'<title>.*</title>')
matches = pattern.finditer(html)
for match in matches:
    print(match)
```

```python
"""to display values only use groups"""
pattern = re.compile(r'<title>(.*)</title>')
matches = pattern.finditer(html)
for match in matches:
#     print(dir(match))
    print(match.group(1))
```

```python
"""Find firstname and last name"""
pattern = re.compile(r'<li>([a-zA-Z]+) ([a-zA-Z]+)<br/>')
matches = pattern.finditer(html)
for match in matches:
#     print(dir(match))
#     print(match.group(0))
#     print(match.group(1))
    print(match.group(2))
```

```python
"""First Name starting with a particular letter"""
pattern = re.compile(r'<li>(C[a-zA-Z]+) (\w+)<br/>')
matches = pattern.finditer(html)
for match in matches:
    print(match.group(1),match.group(2))
```

```python
"""Phone Numbers"""
pattern = re.compile(r'<br/>(\(\d{3}\) \d{3}-\d{4})</li>')
matches = pattern.finditer(html)
for match in matches:
    print(match.group(1))
```

```python
"""Full Address"""
pattern = re.compile(r'<br/>(.*)<br/>(.*)<br/>')
matches = pattern.finditer(html)
for match in matches:
    print(match.group(1), match.group(2))
```

```python
"""Inside tags"""
pattern = re.compile(r'<h.>.*</h.>')
matches = pattern.finditer(html)
for match in matches:
    print(match)
```
