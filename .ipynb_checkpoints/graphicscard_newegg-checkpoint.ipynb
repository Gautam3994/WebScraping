```python
from bs4 import BeautifulSoup
import requests
import re
import csv

with open('ram.csv', 'w', newline="") as csv_file:
# with open('graphics_card.csv', 'a', newline="") as csv_file:
    headers = ["Product_Name", "Brand", "Price", "Rating"]
    writer_ = csv.writer(csv_file)
    writer_.writerow(headers)
#     for i in range(14):
    for i in range(49):
        url = f"https://www.newegg.com/global/in-en/p/pl?d=ram&page={i+1}"
#             url = f"https://www.newegg.com/global/in-en/p/pl?d=graphics%20card&Page={i+1}&Order=BESTMATCH"
        try:
            response = requests.get(url)
        except:
            print(response.status_code)
            pass
        else:
            soup = BeautifulSoup(response.text, "lxml")
            pattern = re.compile(r"[0-9,]+")
            grid = soup.find("div", class_="items-view is-grid")
            for container in grid.find_all("div", class_="item-container"):
#     if "RTX 20" in container.img["title"]:
                product_name = container.img["title"]
                try:
                    brand_name = container.find("div", class_="item-branding").img["title"]
                except:
                    brand_name = "Brand Name Not Available"
                current_price_matches = pattern.finditer(container.find("li", class_="price-current").text)
                try:
                    for match in current_price_matches:
                        price = match.group(0)
                        break
                except:
                    price = "Price not specified"
                try:
                    rating = container.find("div", class_="item-branding").find("a", class_="item-rating")["title"][-1]
                except:
                    rating = f"No rating"
                writer_.writerow([product_name, brand_name, price, rating])
            print("Success for: ", response.url)
print("Success")
```
