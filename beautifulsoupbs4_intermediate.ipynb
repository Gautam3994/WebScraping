```python
"""Navigating sideways - all of this can be imitated using find as well"""
"""next_siblings"""
with open("files/Tom_Jerry.html") as html:
    soup =  BeautifulSoup(html, 'lxml')
# print(soup.prettify())
print(soup.a.next_sibling)
print(soup.b.next_sibling)
print(soup.b.next_sibling.next_sibling)
"""The last sibling wont have any next sibling"""
print(soup.c.next_sibling)
print("------------------")
"""To get all the upcoming siblings use next_siblings"""
for sibling in soup.b.next_siblings:
    print(sibling)
```

```python
"""previous sibling"""
print(soup.c.previous_sibling)
print("-------------------")
for sibling in soup.c.previous_siblings:
    print(sibling)
print(soup.a.previous_sibling)
```

```python
"""next element"""
print(soup.prettify())
print("-----------------------")
print(soup.b.next_element)
print(soup.b.next_element.next_element)
print(soup.b.next_element.next_element.next_element)
print("-------------------------------")
for a in soup.b.next_elements:
    print("-", a)
```

```python
"""Using custom function"""
"""This function returns all tags with a src and not a href"""
def has_src_but_no_href(tag):
    return tag.has_attr('src') and not tag.has_attr('href')
soup.find_all(has_src_but_no_href)
```

```python
"""This function returns all tags with a href and not a src"""
def has_href_but_no_src(tag):
    return tag.has_attr('href') and not tag.has_attr('src')
soup.find_all(has_href_but_no_src)
```

```python
"""This function returns all tags with a href and 'eve' in the value of the href"""
def eve_apple(href):
    return href and re.compile("eve").search(href)
soup.find_all(href = eve_apple )
```

```python
"""This functions returns all tags with href that does not point to wikipedia"""
def not_wiki(href):
    return  href and not re.compile("wiki").search(href)
soup.find_all(href=not_wiki)
```

```python
"""Getting relative links and absolute links"""
"https://www.scrapehero.com/how-to-fake-and-rotate-user-agents-using-python-3/"
url = "https://www.amazon.com/"
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
source = requests.get(url, headers={'User-Agent': user_agent}).text
with open("amazon.html", "w", encoding="utf-8") as html:
    html.write(source)
soup = BeautifulSoup(source, "lxml")
print(soup.prettify())
```

```python
"""to find all links"""
for a in soup.find_all("a", href= re.compile(r"^http")):
    print(a["href"])
```

```python
"""find all  link relative and absolute"""
for a in soup.find_all('a', href=True):
    print(a["href"])
```

```python
"""convert relative to absolute"""
for a in soup.find_all('a', href=True):
    if not a["href"].startswith("http"):
        link = url + a["href"]
    else:
        link = a["href"]
    print(link)
```

```python
"""a link tags"""
for a in soup.find_all(["a", "link"], href= re.compile(r"^http")):
    print(a["href"])
```

```python
try:
    for img in soup.find_all("img"):
        if not img["src"].startswith("http"):
            link = url + img["src"]
        else:
            link = img["src"]
        print(link)
except:
    pass
```
